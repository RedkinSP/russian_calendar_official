```python
from datetime import date, datetime, timedelta
import holidays
import pandas as pd
```

```python
ru_holidays = holidays.RU()
```

```python
ru_holidays
```

```python
ru_holidays.get('2023-01-01')
```

```python
def holidays_list(base=date(datetime.now().year, 1, 1) - timedelta(days=365*2), years=3):
    date_list = [base + timedelta(days=x) for x in range(years * 365)]
    date_status = [x in ru_holidays for x in date_list]
    date_desc = [ru_holidays.get(x) for x in date_list]
    return zip(date_list, date_status, date_desc)
```

```python
df = pd.DataFrame(holidays_list(), columns=['Date', 'Status', 'Description'])
```
